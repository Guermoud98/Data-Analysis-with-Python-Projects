```python
# converting the duration_ms to seconds

df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
df_tracks = df_tracks.reset_index(drop=True)
df_tracks
```

```python
corr_df = df_tracks.drop(["key", "mode", "explicit"],axis = 1).corr()
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df,annot= True, cmap="inferno", vmin=-1, vmax=1, linecolor ="Black", linewidth = 1)
heatmap.set_title("Correlation Heatmap Between Variable")
```
