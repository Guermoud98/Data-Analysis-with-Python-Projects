```python
import pandas as pd
```

```python
# load the data into a Pandas dataframe
path = "C:\\Users\Administrateur\\Data-Analysis-with-Python-Projects\\DataAnalysis\\Global Emissions\\Methane_final.csv"
df = pd.read_csv(path)
df
```

```python
df["country"].unique()
```

```python
# Which country has the highest methane emissions overall?
max_emission = df["emissions"].max()
max_emission_by_country = df.loc[df["emissions"] == max_emission, "country"].values[0]
print("the country which has the highest methane emissions overall is", max_emission_by_country)
```

```python
df
```

```python
# Which type is responsible for the majority of methane emissions?
max_emission = df["emissions"].max()
sector_majority_emissions = df.loc[df["emissions"] == max_emission, "type"].values[0]
print("the type which is responsible for the majority of methane emissions is: ",sector_majority_emissions)
```

```python
# What is the total methane emissions for each country in the dataset?
total_emi_by_country = df.groupby("country").sum().drop('Unnamed: 0', axis=1)
total_emi_by_country
```

```python
# What is the average methane emissions for each type in the dataset?
average_emi_by_type = df.groupby("type").mean().drop('Unnamed: 0', axis=1)
average_emi_by_type
```

```python
# What is the maximum methane emissions for each country in the dataset?
max = df.groupby("country").max()
max[["emissions"]]
```

```python
# What is the minimum methane emissions for each country in the dataset?
min = df.groupby("country").min()
min[["emissions"]]
```

```python
# What is the percentage contribution of each type to the total methane emissions?
total_emission= df["emissions"].sum()
emission_by_sector = df.groupby("type").sum().drop('Unnamed: 0', axis=1)
contribution = (total_emission/emission_by_sector) * 100
contribution
```
